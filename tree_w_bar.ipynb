```python
%%R
library(phyloseq)
library(ggplot2)
library(ape)
library(dplyr)
library(grid)
library(magrittr)
```

```python
%%R
tree = ladderize(read.tree("RAxML.tree"))
```

```python
%%R
tree_dfs = phyloseq:::tree_layout(tree)
edgeDF = data.frame(tree_dfs$edgeDT)
vertDF = data.frame(tree_dfs$vertDT)
```

```python
%%R
taxDF = read.table("color.branches.csv", 
                   sep = ",",
                   header = TRUE,
                   stringsAsFactors = FALSE)
edgeDF = left_join(data.frame(edgeDF), taxDF, by = c("OTU" = "FULLID"))
```

```python
%%R
loopDF = read.table("color.hasLoop.new.csv",
                    sep = ",")
names(loopDF) = c("FULLID", "loop")
edgeDF = left_join(edgeDF, data.frame(loopDF), by = c("OTU" = "FULLID"))
```

```python
%%R
color_df = taxDF %>% group_by(Phylum) %>% summarize(color = first(color))
```

```python
%%R
colors = color_df$color
names(colors) = color_df$Phylum
colors
```

```python
%%R
names.df = data.frame(OTU = c("gi|18143804|dbj|BAB79853.1|", 
                              "gi|494702903|ref|YP_007968364.1|",
                              "gi|312207879|pdb|3LPG|A",
                              "gi|114939|sp|P00722.2|BGAL_ECOLI"),
                      species = c("Clostridium perfringens str. 13", 
                                  "Streptococcus agalactiae", 
                                  "Escherichia coli K-12",
                                  "Escherichia coli K-12\nbeta-galactosidase"))
names.df$tree.id = with(names.df, paste(species, OTU, sep = "\n"))
edgeDF = left_join(edgeDF, names.df)
```

```python
%%R -w 600 -h 250
count.df = read.table("uniqueHitsInSPADES_ALLpeople.bs200.counts")
names(count.df) = c("OTU", "count")

count.df %<>% left_join(edgeDF)

p = ggplot(count.df, aes(x = y, y = count))

p = p + geom_bar(stat = "identity")

p
```

```python
%%R -h 1500 -w 900
xmax = max(edgeDF$xright)
ymin = min(edgeDF$y)
ymax = max(edgeDF$y)
max.count = max(count.df$count)
buffer = 0.35
small.buffer = 0.025
count.scale = 20

p = ggplot()

p = p + geom_segment(data = edgeDF, aes(x = xleft, xend = xright, y = y, yend = y))
#p = p + geom_segment(data = edgeDF[!is.na(edgeDF$OTU),], size = 1.5,
#                     aes(x = xleft, xend = xright, y = y, yend = y, color = Phylum))
p = p + geom_segment(data = vertDF, aes(x = x, xend = x, y = vmin, yend = vmax))

p = p + geom_segment(aes(y = 30, yend = 30, x = 1, xend = 1.1), size = 2)
p = p + annotate("text", label = "0.1 subst/site", x = 1.05, y = 32, hjust = 0.5, size = 4.5)

p = p + geom_rect(data = edgeDF[!is.na(edgeDF$OTU),], size = 2.5, shape = 21, color = NA, alpha = 0.65,
                   aes(xmin = xright + 0.05, xmax = xmax + buffer, ymin = y + 0.5, ymax = y - 0.5, fill = Phylum))

p = p + geom_point(data = edgeDF %>% filter(label >= 60, !is.na(label)), 
                   aes(x = xright, y = y), size = 5)

p = p + geom_rect(data = count.df, size = 1, shape = 21, fill = "black", color = "white", alpha = 0.65,
                   aes(xmin = xmax + buffer + small.buffer, 
                       xmax = xmax + buffer + small.buffer + count / count.scale, 
                       ymin = y + 0.5, 
                       ymax = y - 0.5))
p = p + geom_segment(color = "grey50",
                     aes(x = xmax + buffer + small.buffer,
                     xend = xmax + buffer + small.buffer,
                     y = ymin - 0.5,
                     yend = ymax + 0.5))
p = p + geom_segment(color = "grey50",
                     aes(x = xmax + buffer + small.buffer + max.count / (2 * count.scale),
                     xend = xmax + buffer + small.buffer + max.count / (2 * count.scale),
                     y = ymin - 0.5,
                     yend = ymax + 0.5))
p = p + geom_segment(color = "grey50", 
                     aes(x = xmax + buffer + small.buffer + max.count / count.scale,
                     xend = xmax + buffer + small.buffer + max.count / count.scale,
                     y = ymin - 0.5,
                     yend = ymax + 0.5))
p = p + annotate("text", label = "0",
                 x = xmax + buffer + small.buffer,
                 y = ymax, 
                 vjust = -0.5,
                 hjust = 0.5)
p = p + annotate("text", label = as.character(max.count / 2),
                 x = xmax + buffer + small.buffer + max.count / (2 * count.scale),
                 y = ymax, 
                 vjust = -0.5,
                 hjust = 0.5)
p = p + annotate("text", label = as.character(max.count), 
                 x = xmax + buffer + small.buffer + max.count / count.scale,
                 y = ymax, 
                 vjust = -0.5, 
                 hjust = 0.5)
p = p + annotate("text", label = "Count",
                 x = xmax + buffer + small.buffer + max.count / (2 * count.scale),
                 y = ymax, 
                 vjust = -2.5,
                 hjust = 0.5)

#p = p + annotate("text", label = "N")

#p = p + geom_text(data = edgeDF, vjust = 0.5, hjust = 0, 
#                  fontface = 3,
#                  size = 4,
#                  lineheight = 1,
#                  aes(x = xmax + buffer + small.buffer, y = y, label = tree.id))
#p = p + geom_segment(data = edgeDF[!is.na(edgeDF$species),], 
#                     arrow = arrow(length = unit(0.25, "cm")),
#                     size = 1,
#                     aes(xend = xright + small.buffer, x = xmax + buffer + small.buffer, y = y, yend = y))

p = p + scale_fill_manual(values = colors)

#p = p + scale_x_continuous(limits = c(0, 7))

#p = p + scale_fill_manual(values = c("#fc8d62", "#8da0cb"))

p = p + theme_bw()

p = p + theme(axis.text = element_blank(),
              axis.title = element_blank(),
              axis.ticks = element_blank(),
              panel.grid = element_blank(),
              panel.border = element_blank(),
              legend.title = element_text(size = 18),
              legend.text = element_text(size = 18),
              legend.position = c(0.175, 0.4),
              legend.box.just = "left")

p
```
