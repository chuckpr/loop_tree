###Load R packages we'll use

```python
%%R
library(phyloseq)
library(ggplot2)
library(ape)
library(dplyr)
```

###Read in the tree

```python
%%R
tree = read.tree("FinalTrimmedTreeBasedon0.8GetSeqs.NKYWeirdosRemoved.8.phylip_phyml_tree.txt")
```

```python
%%R
tree.rooted = ladderize(root(tree, "gi|114939|sp|P00722.2|BGAL_ECOLI"))
```

###Translate tree to "edge" and "vert" dataframes using the "tree_layout" function withing Phyloseq (originally from ggphylo)

```python
%%R
tree_dfs = phyloseq:::tree_layout(tree.rooted)
edgeDF = tree_dfs$edgeDT
vertDF = tree_dfs$vertDT
```

###Load the tip data

```python
%%R
taxDF = read.table("color.branches.csv", 
                   sep = ",",
                   header = TRUE,
                   stringsAsFactors = FALSE)
edgeDF = left_join(data.frame(edgeDF), taxDF, by = c("OTU" = "FULLID"))
```

```python
%%R
loopDF = read.table("color.hasLoop.csv",
                    sep = ",")
names(loopDF) = c("FULLID", "loop")
edgeDF = left_join(edgeDF, loopDF, by = c("OTU" = "FULLID"))
```

```python
%%R
color_df = taxDF %>% group_by(Phylum) %>% summarize(color = first(color))
```

```python
%%R
colors = color_df$color
names(colors) = color_df$Phylum
colors
```

###Voila!

```python
%%R
colnames(edgeDF)
```

```python
%%R -h 1200
p = ggplot()

p = p + geom_segment(data = edgeDF, aes(x = xleft, xend = xright, y = y, yend = y))
p = p + geom_segment(data = edgeDF[!is.na(edgeDF$OTU),], size = 1.2,
                     aes(x = xleft, xend = xright, y = y, yend = y, color = Phylum))
p = p + geom_segment(data = vertDF, aes(x = x, xend = x, y = vmin, yend = vmax))

p = p + geom_rect(data = edgeDF[!is.na(edgeDF$OTU),], size = 2.5, shape = 21, color = NA, alpha = 0.25,
                   aes(xmin = xright + 0.2, xmax = max(xright) + 0.25, ymin = y + 0.5, ymax = y - 0.5, fill = loop))

p = p + scale_color_manual(values = colors)

p = p + theme_bw()

p = p + theme(axis.text = element_blank(),
              axis.title = element_blank(),
              axis.ticks = element_blank(),
              panel.grid = element_blank(),
              panel.border = element_blank(),
              legend.title = element_text(size = 14),
              legend.text = element_text(size = 12))

p
```

```python
%%R -w 800 -h 800
p = p + coord_polar(theta = "y")

p = p + theme(axis.text = element_blank(),
              axis.title = element_blank(),
              axis.ticks = element_blank(),
              panel.grid = element_blank(),
              panel.border = element_blank(),
              legend.title = element_text(size = 14),
              legend.text = element_text(size = 12))

p
```
