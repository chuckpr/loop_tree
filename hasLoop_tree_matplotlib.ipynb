###Bring R data into Python

```python
edgeDF = com.load_data("edgeDF")
vertDF = com.load_data("vertDF")
```

###Rectangular tree

```python
edge_segments = edgeDF.apply(lambda row: [(row["xleft"], row["y"]), (row["xright"], row["y"])], axis = 1).values
vert_segments = vertDF.apply(lambda row: [(row["x"], row["vmin"]), (row["x"], row["vmax"])], axis = 1).values
```

```python
tipDF = edgeDF[edgeDF["OTU"].notnull()]
tip_segments = tipDF.apply(lambda row: [(row["xleft"], row["y"]), (row["xright"], row["y"])], axis = 1).values
```

```python
tip_colors = tipDF.apply(lambda row: (row["red"]/255, row["green"]/255, row["blue"]/255), axis = 1)
```

```python
max_right = tipDF["xright"].max() * 1.05
buff = 0.05
verts = tipDF.apply(lambda row: [(row["xright"] + buff, row["y"] - 0.5), 
                                 (row["xright"] + buff, row["y"] + 0.5),
                                 (max_right, row["y"] + 0.5), 
                                 (max_right, row["y"] - 0.5)], 
            axis = 1).values
```

```python
fig = plt.figure()
fig.set_size_inches([6,18])
ax = fig.add_subplot(111, polar=False)

edge_collection = collections.LineCollection(segments = edge_segments, color="k", linewidths = 1.5)
ax.add_collection(edge_collection)

vert_collection = collections.LineCollection(segments = vert_segments, color="k", linewidths = 1.5)
ax.add_collection(vert_collection)

tip_collection = collections.LineCollection(segments = tip_segments, colors=tip_colors, linewidths=2.5)
ax.add_collection(tip_collection)

loop_collection = collections.PolyCollection(verts, 
                                             facecolors = tipDF.apply(lambda x: "red" if x["loop"] == "hasLoop" 
                                                                      else "blue", axis = 1),
                                             alpha = 0.25)
ax.add_collection(loop_collection)

ax.set_xlim(edgeDF["xleft"].min() - edgeDF["xright"].max() * 0.05, edgeDF["xright"].max() * 1.05)
ax.set_ylim(edgeDF["y"].min() - edgeDF["y"].max() * 0.05, edgeDF["y"].max() * 1.05)

ax.set_xticks([])
ax.set_yticks([])
ax.set_frame_on(False)
```

###Circular tree

```python
max_y = edgeDF["y"].max()
edge_segments = edgeDF.apply(lambda row: [(row["y"] / max_y * 2 * np.pi, row["xleft"]), 
                                          (row["y"] / max_y * 2 * np.pi, row["xright"])], axis = 1).values
```

```python
max_x = vertDF["vmax"].max()

def interp_verts(row):
    verts = np.linspace(row["vmin"] / max_x * 2 * np.pi, 
                        row["vmax"] / max_x * 2 * np.pi, 100)
    return [(i, row["x"]) for i in verts]

vert_segments = vertDF.apply(interp_verts, axis = 1).values
```

```python
tipDF = edgeDF[edgeDF["OTU"].notnull()]
tip_segments = tipDF.apply(lambda row: [(row["y"] / max_y * 2 * np.pi, row["xleft"]), 
                                        (row["y"] / max_y * 2 * np.pi, row["xright"])], axis = 1).values
```

```python
max_right = tipDF["xright"].max() * 1.05
buff = 0.10
verts = tipDF.apply(lambda row: [((row["y"] - 0.5) / max_y * 2 * np.pi, row["xright"] + buff), 
                                 ((row["y"] + 0.5) / max_y * 2 * np.pi, row["xright"] + buff),
                                 ((row["y"] + 0.5) / max_y * 2 * np.pi, max_right), 
                                 ((row["y"] - 0.5) / max_y * 2 * np.pi, max_right)], 
            axis = 1).values
```

```python
fig = plt.figure()
fig.set_size_inches([12,12])
ax = fig.add_subplot(111, polar=True)
ax.set_rmax(max_right)

edge_collection = collections.LineCollection(segments = edge_segments, color="k", linewidths = 2)
ax.add_collection(edge_collection)

vert_collection = collections.LineCollection(segments = vert_segments, color="k", linewidths = 2)
ax.add_collection(vert_collection)

tip_collection = collections.LineCollection(segments = tip_segments, colors=tip_colors, linewidths=3)
ax.add_collection(tip_collection)

loop_collection = collections.PolyCollection(verts, 
                                             facecolors = tipDF.apply(lambda x: "red" if x["loop"] == "hasLoop" 
                                                                      else "blue", axis = 1),
                                             alpha = 0.25)
ax.add_collection(loop_collection)

ax.set_xticks([])
ax.set_yticks([])
ax.set_frame_on(False)
```
