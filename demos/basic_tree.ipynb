###Load R packages we'll use

```python
%%R
library(phyloseq)
library(ggplot2)
library(ape)
library(dplyr)
```

###Read in the tree

```python
%%R
tree = read.tree("FinalTrimmedTreeBasedon0.8GetSeqs.NKYWeirdosRemoved.8.phylip_phyml_tree.txt")
```

###Translate tree to "edge" and "vert" dataframes using the "tree_layout" function withing Phyloseq (originally from ggphylo)

```python
%%R
tree_dfs = phyloseq:::tree_layout(tree)
```

###Voila!

```python
%%R
p = ggplot()

p = p + geom_segment(data = tree_dfs$edgeDT, aes(x = xleft, xend = xright, y = y, yend = y))
p = p + geom_segment(data = tree_dfs$vertDT, aes(x = x, xend = x, y = vmin, yend = vmax))

p = p + theme_bw()

p
```

```python
%%R
p + coord_polar(theta = "y")
```

###You can also add labels to the tips like so (for instance)...

```python
%%R -h 800 -w 800
p = ggplot()

p = p + geom_segment(data = tree_dfs$edgeDT, aes(x = xleft, xend = xright, y = y, yend = y))
p = p + geom_segment(data = tree_dfs$vertDT, aes(x = x, xend = x, y = vmin, yend = vmax))

d = tree_dfs$edgeDT

p = p + geom_text(data = d[!is.na(d$OTU),], 
                  size = 1.5, 
                  hjust = 0,
                  aes(x = 4.25, y = y, angle = y / max(y) * -360 + 90, label = OTU))

p = p + xlim(-0.5, 5)

p = p + theme_bw()

p = p + theme(axis.text = element_blank())

p + coord_polar(theta = "y")
```
